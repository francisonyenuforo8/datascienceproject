```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
from matplotlib import pyplot as plt 
from sklearn import tree
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
dataset.drop(columns = ['customerID'], inplace=True)
dataset.head(14)
```

```python
dataset.columns
```

```python
gender = {"Male": 0, "Female": 1}

dataset['gender'] = dataset['gender'].map(gender)
```

```python
Churn = {"Yes": 0, "No": 1}

dataset['Churn'] = dataset['Churn'].map(Churn)
```

```python
dataset.head()
```

```python
X = dataset[['gender','MonthlyCharges','SeniorCitizen']]
y = pd.Categorical(dataset['Churn'])
```

```python
X.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X_train,y_train)
predictions = dt.predict(X_test)
```

```python
y_test
```

```python
# Accuracy
accuracy_score(y_test, predictions)
```

```python
# Classification report
print(f'classification_report: \n{classification_report(y_test, predictions)}')
```
